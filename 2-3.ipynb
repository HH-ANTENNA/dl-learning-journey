```python
x = torch.arange(12, dtype=torch.float32).reshape(2, 2, 3)
print(x)
# 默认行为：计算整个张量的L2范数（标量）
a1 = torch.linalg.norm(x)
print(a1)

# 指定维度：在第0，第1和第3上计算L2范数
y0 = torch.linalg.norm(x, dim=0)
y1 = torch.linalg.norm(x, dim=1)
y2 = torch.linalg.norm(x, dim=2)
print(y0,y0.shape)
print(y1,y1.shape)
print(y2,y2.shape)


# 计算L1范数
# 在第0，第1和第2轴上计算L1范数
l1_0 = torch.linalg.norm(x,ord=1,dim=0)
l1_1 = torch.linalg.norm(x,ord=1,dim=1)
l1_2 = torch.linalg.norm(x,ord=1,dim=2)
print(l1_0)
print(l1_1)
print(l1_2)
```

练习8
torch.linalg.norm(x,ord=1,dim=0)  ord默认=2 dim默认None(计算整个张量的范数)
ord = 1 计算对应的L1范数
